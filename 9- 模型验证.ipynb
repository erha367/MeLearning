### 交叉验证

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
```

```python
wine = load_wine()
svc = SVC(kernel='linear')
scores = cross_val_score(svc,wine.data,wine.target,cv=6)
scores,scores.mean()
```

### 随机拆分

```python
from sklearn.model_selection import ShuffleSplit
```

```python
shuffle_split = ShuffleSplit(test_size=.2,train_size=.7,n_splits=10)
scores = cross_val_score(svc,wine.data,wine.target,cv=shuffle_split)
scores
```

### 挨个试试

```python
from sklearn.model_selection import LeaveOneOut
```

```python
cv = LeaveOneOut()
scores = cross_val_score(svc,wine.data,wine.target,cv=cv)
len(scores),scores.mean()
```

### 网格搜索优化模型参数

```python
from sklearn.linear_model import Lasso 
from sklearn.model_selection import train_test_split
```

```python
#random_state = 0 分数0.82
X_train,X_test,y_train,y_test = train_test_split(wine.data,wine.target,random_state=38)
best_score = 0
for alpha in [0.01,0.1,1,10]:
    for max_iter in [100,1000,5000,10000]:
        lasso = Lasso(alpha=alpha,max_iter=max_iter)
        lasso.fit(X_train,y_train)
        score = lasso.score(X_test,y_test)
        if score > best_score:
            best_score = score
            best_parameters = [alpha,max_iter]
print(best_score,best_parameters)            
```

### 交叉验证结合网格搜索

```python
import numpy as np
```

```python
best_score = 0
for alpha in [0.01,0.1,1,10]:
    for max_iter in [100,1000,5000,10000]:
        lasso = Lasso(alpha=alpha,max_iter=max_iter)
        scores = cross_val_score(lasso,X_train,y_train,cv=6)
        score = np.mean(scores)
        if score > best_score:
            best_score = score
            best_parameters = [alpha,max_iter]
print(best_score,best_parameters)  
```

### 分类模型的预测准确率

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X,y = make_blobs(n_samples=200,random_state=1,centers=2,cluster_std=5)
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.autumn,edgecolors='k')
plt.show()
```

```python
from sklearn.naive_bayes import  GaussianNB
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=68)
gnb = GaussianNB()
gnb.fit(X_train,y_train)
predict_proba = gnb.predict_proba(X_test)
#准确率形态
predict_proba.shape,predict_proba[:1]
```

```python
x_min,x_max = X[:,0].min()-1,X[:,0].max()+1
y_min,y_max = X[:,1].min()-1,X[:,1].max()+1
    #meshgrid函数用两个坐标轴上的点在平面上画网格
xx,yy = np.meshgrid(np.arange(x_min,x_max,.02),np.arange(y_min,y_max,.02))
    #预测,ravel()将多维数组降位一维,c_将切片对象转换为沿第二轴的连接
z = gnb.predict_proba(np.c_[xx.ravel(),yy.ravel()])[:,1]
z = z.reshape(xx.shape)
    #绘制分类图
plt.contourf(xx,yy,z,cmap=plt.cm.Pastel1,alpha=.8)
    #绘制散点图
plt.scatter(X_train[:,0],X_train[:,1],c=y_train,cmap=plt.cm.spring,edgecolor='k')
plt.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap=plt.cm.spring,edgecolor='k',alpha=0.6)
plt.xlim(xx.min(),xx.max())
plt.ylim(yy.min(),yy.max())
plt.xticks(())
plt.yticks(())
plt.title('GaussNB')
plt.show()
```

```python
### 决定系数
```

```python
svc = SVC().fit(X_train,y_train)
dec_func = svc.decision_function(X_test)
#打印前5个，正的是分类1，负的是分类2
dec_func[:5]
```
