```python
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
import numpy as np
```

```python
blobs = make_blobs(random_state=1,centers=1)
```

```python
X_blobs = blobs[0]
plt.scatter(X_blobs[:,0],X_blobs[:,1],c='r',edgecolor='k')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_blobs)
```

```python
def draw_mlp_pic(X_blobs,mlp):
    x_min,x_max = X_blobs[:,0].min()-1,X_blobs[:,0].max()+1
    y_min,y_max = X_blobs[:,1].min()-1,X_blobs[:,1].max()+1
    #meshgrid函数用两个坐标轴上的点在平面上画网格
    xx,yy = np.meshgrid(np.arange(x_min,x_max,.02),np.arange(y_min,y_max,.02))
    #预测,ravel()将多维数组降位一维,c_将切片对象转换为沿第二轴的连接
    z = mlp.predict(np.c_[xx.ravel(),yy.ravel()])
    z = z.reshape(xx.shape)
    plt.figure(1)
    plt.clf()
    plt.imshow(z,interpolation='nearest',extent=[xx.min(),xx.max(),yy.min(),yy.max()],cmap=plt.cm.summer,aspect='auto',origin='lower')
    plt.plot(X_blobs[:,0],X_blobs[:,1],'r.',markersize=5)
    centoids = kmeans.cluster_centers_
    plt.scatter(centoids[:,0],centoids[:,1],marker='x',s=150,linewidths=3,color='b',zorder=10)
    plt.xlim(x_min,x_max)
    plt.ylim(y_min,y_max)
    plt.xticks(())
    plt.yticks(())
    plt.show()
```

```python
draw_mlp_pic(X_blobs,kmeans)
```

```python
### 凝聚聚类
```

```python
from scipy.cluster.hierarchy import dendrogram,ward
```

```python
linkage = ward(X_blobs)
dendrogram(linkage)
ax = plt.gca()
plt.xlabel('Sample index')
plt.ylabel('Cluster distance')
plt.show()
```

```python
### DBSCAN 算法
```

```python
from sklearn.cluster import DBSCAN
```

```python
# eps 越大覆盖的越多，默认0.5
# min_samples越大，核心数据点越少，噪点越多
db = DBSCAN(min_samples=20)
clusters = db.fit_predict(X_blobs)
plt.scatter(X_blobs[:,0],X_blobs[:,1],c=clusters,cmap=plt.cm.cool,s=60,edgecolors='k')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.show()
```
