```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```python
#生成50个点，分类2，方差为6
X,y = make_blobs(n_samples=50,centers=2,random_state=6)
```

```python
#线性内核，L1范数最小化
clf = svm.SVC(kernel='linear',C=1000)
clf.fit(X,y)
```

```python
def support_side(X,y,clf):
    plt.scatter(X[:,0],X[:,1],c=y,s=30,cmap=plt.cm.Paired)
    #建立图像坐标
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    #生成等差数列
    xx = np.linspace(xlim[0],xlim[1],30)
    yy = np.linspace(ylim[0],ylim[1],30)
    YY,XX = np.meshgrid(yy,xx)
    xy = np.vstack([XX.ravel(),YY.ravel()]).T
    Z = clf.decision_function(xy).reshape(XX.shape)
    #分类边界
    ax.contour(XX,YY,Z,colors='k',levels=[-1,0,1],alpha=0.5,linestyles=['--','-','--'])
    #ax.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s=100,linewidth=1,facecolors='none')
    plt.show()
```

```python
support_side(X,y,clf)
```

```python
#RBF内核，径向基内核  
# -- gamma越小，RBF的半径越大，有更多的点被圈进来，决策边界平滑，模型简单（趋向于欠拟合）
# -- C越小，模型越受到限制（欠拟合）
clf = svm.SVC(kernel='rbf',gamma=0.7,C=1000)
clf.fit(X,y)
```

```python
support_side(X,y,clf)
```

```python
clf = svm.SVC(kernel='poly',degree=3,C=1000)
clf.fit(X,y)
```

```python
support_side(X,y,clf)
```

```python
#L2范数最小化，C较大时，边界严格
clf = svm.LinearSVC(C=1000)
clf.fit(X,y)
```

```python
support_side(X,y,clf)
```

```python
#波士顿房价预测
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
#boston.keys()
#boston.DESCR
#boston.target
```

```python
X,y = boston.data,boston.target
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=8)
X_train.shape,X_test.shape
```

```python
plt.plot(X.min(axis=0),'v',label='min')
plt.plot(X.max(axis=0),'^',label='max')
plt.yscale('log')
plt.legend(loc='best')
plt.xlabel('feaures')
plt.ylabel('feature magnitude')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
plt.plot(X_train_scaled.min(axis=0),'v',label='train_min')
plt.plot(X_train_scaled.max(axis=0),'^',label='train_max')
plt.plot(X_test_scaled.min(axis=0),'v',label='test_min')
plt.plot(X_test_scaled.max(axis=0),'^',label='test_max')
#plt.yscale('log')
plt.legend(loc='best')
plt.xlabel('scaled feaures')
plt.ylabel('scaled feature magnitude')
plt.show()
```

```python
svr = svm.SVR(C=100,gamma=0.1)
```

```python
svr.fit(X_train_scaled,y_train)
```

```python
svr.score(X_test_scaled,y_test)
```
