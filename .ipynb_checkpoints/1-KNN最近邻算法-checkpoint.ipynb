```python
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
#样本数200，分类为2的数据集 make_blobs,聚类数据生成器
data = make_blobs(n_samples=200,centers=2,random_state=8)
```

```python
x,y = data
y
```

```python
#数据可视化，散点图
plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.winter,edgecolor='k')
plt.show()
```

```python
#训练器，训练数据
clf = KNeighborsClassifier()
clf.fit(x,y)
```

```python
def draw_pic(x,y,clf):
    x_min,x_max = x[:,0].min()-1,x[:,0].max()+1
    y_min,y_max = x[:,1].min()-1,x[:,1].max()+1
    #meshgrid函数用两个坐标轴上的点在平面上画网格
    xx,yy = np.meshgrid(np.arange(x_min,x_max,.02),np.arange(y_min,y_max,.02))
    #预测,ravel()将多维数组降位一维,c_将切片对象转换为沿第二轴的连接
    z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
    z = z.reshape(xx.shape)
    #绘制分类图
    plt.pcolormesh(xx,yy,z,cmap=plt.cm.Pastel1)
    #绘制散点图
    plt.scatter(x[:,0],x[:,1],c=y,cmap=plt.cm.spring,edgecolor='k')
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
    plt.title("KNN")
    plt.scatter(6.75,4.82,marker='*',c='green',s=200)
    plt.show()
```

```python
clf.predict([[6.75,4.82]])
```

```python
draw_pic(x,y,clf)
```

```python
#生成5组数据集
data2 = make_blobs(n_samples=500,centers=5,random_state=8)
x2,y2 = data2
clf.fit(x2,y2)
plt.scatter(x2[:,0],x2[:,1],c=y2,cmap=plt.cm.spring,edgecolor='k')
plt.show()
```

```python
#通用画图函数
def draw_pic(x2,y2,clf):
    x_min,x_max = x2[:,0].min()-1,x2[:,0].max()+1
    y_min,y_max = x2[:,1].min()-1,x2[:,1].max()+1
    xx,yy = np.meshgrid(np.arange(x_min,x_max,.02),np.arange(y_min,y_max,.02))
    z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
    z = z.reshape(xx.shape)
    plt.pcolormesh(xx,yy,z,cmap=plt.cm.Pastel1)
    plt.scatter(x2[:,0],x2[:,1],c=y2,cmap=plt.cm.spring,edgecolor='k')
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
    plt.title("KNN-5")
    plt.show()
```

```python
draw_pic(x2,y2,clf)
```

```python
print('模型评分:{:.2f}'.format(clf.score(x2,y2)))
```

```python
# 生成回归模型的数据
from sklearn.datasets import make_regression
#1个特征，50标准差的噪点
X,y = make_regression(n_features=1,n_informative=1,noise=50,random_state=8)
```

```python
plt.scatter(X,y,c='orange',edgecolors='k')
plt.show()
```

```python
#导入回归模型包
from sklearn.neighbors import KNeighborsRegressor
#默认为2
reg = KNeighborsRegressor(n_neighbors=2)
reg.fit(X,y)
```

```python
#在指定的间隔内返回均匀间隔的数字
z = np.linspace(-3,3,200).reshape(-1,1)
#散点图
plt.scatter(X,y,c='red',edgecolors='k')
#预测的折线图
plt.plot(z,reg.predict(z),c='k',linewidth=3)
plt.title('KNN Regressor')
plt.show()
```

```python
print('模型评分:{:.2f}'.format(reg.score(X,y)))
```
